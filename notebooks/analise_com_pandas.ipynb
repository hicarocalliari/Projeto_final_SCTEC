```python
from pathlib import Path
import sys
import zipfile
import pandas as pd

sys.path.insert(0, str(Path.cwd().resolve().parent))

from _MySQL import banco, config

from scripts import extrair
from scripts import transformar
```

```python
try:
    caminho_zip = extrair.localizar_zip()

    dfs = {}

    with zipfile.ZipFile(caminho_zip) as zip_aberto:
        for arquivo in config.ARQUIVOS.values():
            nome_csv = arquivo["csv"]
            nome_df = arquivo["tabela_raw"]

            with zip_aberto.open(nome_csv) as arquivo_csv:
                dfs[nome_df] = pd.read_csv(
                    arquivo_csv,
                    sep=";",
                    encoding="latin1"
                )

    df_pagamento = dfs["raw_pagamento"]
    df_viagem = dfs["raw_viagem"]
    df_passagem = dfs["raw_passagem"]
    df_trecho = dfs["raw_trecho"]

    print("=== DataFrames carregados com sucesso! ===")

except Exception as erro:
    print("[ERRO] Algo deu errado:", erro)
    raise
```
